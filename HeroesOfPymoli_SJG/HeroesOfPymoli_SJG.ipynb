### Heroes Of Pymoli Data Analysis
* Of the 576 active players, the vast majority are male (84.03%). There also exists, a smaller, but notable proportion of female players (14.06%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.58%) and 25-29 (13.37%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
```

## Player Count

```python
#players is the unique screen names 

players = df['SN'].unique()
player_count = len(players)
pc1 = [player_count]

pc_df = pd.DataFrame({'Total Players': pc1})
pc_df
```

## Purchasing Analysis (Total)

```python
## Purchasing Analysis (Total)

#unique items
item_count = len(df['Item ID'].unique())

#item average cost - 145 unique prices and 183 unique items so maybe use a groupby object
cst = df['Price'].unique()

#this average matches the average in the solution, but is NOT the average cost of an item,
#it is the average cost of an item purchased
avg = df['Price'].mean()

#both of these give the total number of purchases
p_tot = df['Purchase ID'].max() + 1

#total revenue
rev_tot = [df['Price'].sum()]

purch_analy = pd.DataFrame({"Number of Unique Items": item_count, 
                            "Average Price": avg,
                            "Number of Purchases": p_tot, 
                            "Total Revenue": rev_tot
                           })



purch_analy
```

## Gender Demographics

```python
m_df = df.loc[df['Gender'] == 'Male']
f_df = df.loc[df['Gender'] == 'Female']
o_df = df.loc[df['Gender'] == 'Other / Non-Disclosed']


m_count = len(m_df['SN'].unique())
f_count= len(f_df['SN'].unique())
o_count = len(o_df['SN'].unique())


m_prcnt = m_count / player_count * 100
f_prcnt = f_count / player_count * 100
o_prcnt = o_count / player_count * 100

counts = [m_count, f_count, o_count]
prcnts = [m_prcnt, f_prcnt, o_prcnt]

indx = ['Male', 'Female', 'Other / Non-Disclosed']
 
gen_dem = pd.DataFrame({
    'Total Count': counts,
    'Percentage of Players': prcnts
})
gen_dem.set_index([pd.Index(indx)])
```


## Purchasing Analysis (Gender)

```python
m_purch_count = len(m_df['Price'])
m_purch_tot = m_df['Price'].sum()
m_avg = m_purch_tot / m_purch_count
m_avg_person = m_purch_tot / m_count

f_purch_count = len(f_df['Price'])
f_purch_tot = f_df['Price'].sum()
f_avg = f_purch_tot / f_purch_count
f_avg_person = f_purch_tot / f_count

o_purch_count = len(o_df['Price'])
o_purch_tot = o_df['Price'].sum()
o_avg = o_purch_tot / o_purch_count
o_avg_person = o_purch_tot / o_count


gen_pcount =[m_purch_count,f_purch_count,o_purch_count]
gen_avg = [m_avg,f_avg,o_avg]
gen_tot = [m_purch_tot,f_purch_tot,o_purch_tot]
gen_avg_person = [m_avg_person,f_avg_person,o_avg_person]

gen_ana_df = pd.DataFrame({
    'Purchase Count': gen_pcount,
    'Average Purchase Price': gen_avg,
    'Total Purchase Value': gen_tot,
    'Avg. Total Purchase per Person': gen_avg_person
})
gen_ana_df = gen_ana_df.set_index([pd.Index(indx)])
gen_ana_df.head(3)
```

## Age Demographics

```python
bins = [0, 9, 14,19,24,29,34,39, df['Age'].max()]
bin_name = ['>10', '10-14','15-19','20-24','25-29','30-34','35-39','40+']
df['Age Group'] = pd.cut(df["Age"], bins, labels=bin_name)

age_gp = df.groupby('Age Group')

age_count = age_gp['SN'].nunique()
age_prcnt = []
for x in age_count:
    age_prcnt.append(x/player_count*100)

age_dem = pd.DataFrame({
    'Total Count': age_count,
    'Percentage of Players': age_prcnt
})
age_dem = age_dem.set_index([pd.Index(bin_name)])

age_dem
```

## Purchasing Analysis (Age)

```python
age_purch_tot = age_gp['Price'].sum()
age_purch_avg = age_gp['Price'].mean()
age_purch_count = age_gp['Price'].count()
age_avg_person = []
i = 0
while i < len(age_purch_tot):
    age_avg_person.append(age_purch_tot[i] / age_count[i] )
    i = i+1


age_analysis_df = pd.DataFrame({
    'Purchase Count': age_purch_count,
    'Average Purchase Price': age_purch_avg,
    'Total Purchase Value': age_purch_tot,
    'Avg. Total Purchase per Person': age_avg_person
})
age_analysis_df = age_analysis_df.set_index([pd.Index(bin_name)])

age_analysis_df
```

## Top Spenders

```python
spend_gp = df.groupby('SN')
numb_purch = spend_gp['Purchase ID'].count()
numb_purch=numb_purch.sort_values(ascending = False)
numb_purch_df = numb_purch.to_frame(name='Purchase Count')

spend_tot = spend_gp['Price'].sum()
spend_tot = spend_tot.sort_values(ascending = False)
spend_tot_df = spend_tot.to_frame(name='Total Purchase Value')


spenders_df = numb_purch_df.join(spend_tot_df)

spenders_df['Average Purchase Price']= spenders_df["Total Purchase Value"]/spenders_df["Purchase Count"]

spenders_df = spenders_df[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]

spenders_df = spenders_df.sort_values('Total Purchase Value', ascending = False)

spenders_df.head()

```

## Most Popular Items

```python
item_df = df[['Item ID', 'Item Name', 'Price']]
item_gp = item_df.groupby(['Item ID', 'Item Name'])

item_purch_count = item_gp['Item Name'].count()
ipc_df = item_purch_count.to_frame(name = 'Purchase Count')

item_df2 = item_df.drop_duplicates()
item_df2.set_index(['Item ID', 'Item Name'])


item_df3 = pd.merge(item_df2,ipc_df, on = 'Item ID')
item_df3 = item_df3.sort_values('Purchase Count', ascending = False)
item_df3['Total Purchase Value'] = item_df3['Price'] * item_df3['Purchase Count']
item_df3 = item_df3.set_index(['Item ID', 'Item Name'])
item_df3 = item_df3[['Purchase Count', 'Price', 'Total Purchase Value']]
item_df3 = item_df3.rename(columns={"Price": "Item Price"})
item_df3.head()

```
